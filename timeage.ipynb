```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
timeage=pd.read_csv("C:/corona/TimeAge.csv")
```

```python
#timeage데이터프레임 미리보기
timeage.head()
```

```python
#timeage 에서 time열 삭제하기 axis=0 행 삭제  axis=1 열 삭제 
timeage=timeage.drop(['time'],axis=1)

# column 이름 영어 --> 한국어 바꾸기
timeage.columns=['날짜','연령대','확진자누적수','사망자누적수']
timeage.head()
```

```python
#연령대 column 의 값 종류 보기
timeage['연령대'].unique()
```

```python
#연령대가 10s 인 값들만 추출
age_10s=timeage.loc[timeage['연령대']=='10s'].reset_index(drop=True)
age_10s.max()
```

```python
#연령대가 10s 인 값들의 날짜가 지남에따른 확진자누적수 변화 
plt.figure(figsize=(20,10))
plt.xticks(rotation=30)
plt.xlabel('날짜')
plt.ylabel('확진자누적수')
sns.lineplot(data=age_10s,x='날짜',y='확진자누적수')
```

```python
#timeage데이터의 연령대별 확진자누적수 변화

plt.figure(figsize=(20,10))
plt.xticks(rotation=30)
plt.xlabel('날짜')
plt.ylabel('확진자누적수')
sns.lineplot(data=timeage,x='날짜',y='확진자누적수',hue='연령대')
```
